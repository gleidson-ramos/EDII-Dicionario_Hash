```python
facebook = {}


def criar_perfil(usuarios, nome, idade, cidade):
  usuarios[nome] = {'idade': idade, 'cidade': cidade, 'amigos': []}
  print(f"Olá {nome}. Bem vindo ao facebook!")


def adicionar_amigos(usuarios, user1, user2):
  if user1 and user2 in usuarios:
    usuarios[user1]['amigos'].append(user2)
    usuarios[user2]['amigos'].append(user1)
    print(f"{user1} e {user2} agora são amigos!")
  else:
    print("Pedimos desculpas, mas não foi possível localizar o usuário.")


def encontrar_perfil(usuarios, nome):
  if nome in usuarios:
    user = usuarios[nome]
    print("Perfil Encontrado!")
    print(f"{nome} tem {user['idade']} anos e reside em {user['cidade']}")
    print(f"Amigos de {nome}: ")

    for amigo in user['amigos']:
      print(f"{amigo}")
  else:
    print(f"Me desculpe, mas {nome} não foi encontrado (a) no facebook.")


def excluir_perfil(usuarios, nome):
  if nome in usuarios:
    del usuarios[nome]
    print(f"Perfil de {nome} excluido do facebook.")
    for user in usuarios.values():
      if nome in user['amigos']:
        user['amigos'].remove(nome)
  else:
    print(f"{nome} não tem perfil no facebook.")


def listar_perfis(usuarios):
  nomes = sorted(usuarios.keys())
  for nome in nomes:
    print(nome)


while True:
  print("\nMenu:")
  print("1. Criar Perfil")
  print("2. Adicionar Amigos")
  print("3. Buscar Perfil")
  print("4. Excluir Perfil")
  print("5. Exibir Usuários")
  print("6. Exibir Facebook Completo")
  print("7. Sair")

  escolha = input("Escolha uma opção: ")
  if escolha == '1':
    nome = input("Informe seu nome: ")
    idade = int(input("Informe sua idade: "))
    cidade = input("Informe sua cidade: ")
    criar_perfil(facebook, nome, idade, cidade)
  elif escolha == '2':
    user1 = input("Informe seu nome: ")
    user2 = input("Informe o nome de seu amigo: ")
    adicionar_amigos(facebook, user1, user2)
  elif escolha == '3':
    nome_busca = input("Informe o nome do usuário: ")
    encontrar_perfil(facebook, nome_busca)
  elif escolha == '4':
    nome = input("Informe o nome do usuário que deseja excluir: ")
    excluir_perfil(facebook, nome)
  elif escolha == '5':
    listar_perfis(facebook)
  elif escolha == '6':
    print(facebook)
  elif escolha == '7':
    break
  else:
    print("Opção inválida. Por favor, escolha uma opção válida.")
```
